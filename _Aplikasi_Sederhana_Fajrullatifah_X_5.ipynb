App Bangun Balok

```python
# Progam menghitung volume balok

#Meminta input dari pengguna
panjang = float(input("masukan panjang volume balok(cm):"))
lebar = float(input("masukan lebar volume balok(cm):"))
tinggi = float(input("masukan tinggi volume balok(cm):"))

#Menghitung volume
volume = panjang * lebar * tinggi

#Menampilkan hasil
print( f"volume balok adalah:{volume} cm^3.")
```
